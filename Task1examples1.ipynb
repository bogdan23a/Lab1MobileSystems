## Section 9.8 Generating sampled sine-wave
Now create and plots a list y containing 80 samples of a sine-wave.  The sine-wave has amplitude 1.0 volts.  Its frequency is 4.7 cycles per second (Hz) and it is sampled at Fs =100 samples per second (Hz).

```python
# Cell 1 for creating & plotting a sine-wave using lists.
import numpy
from matplotlib import pyplot
% matplotlib inline
Fs = 100.0   # Sampling frequency (Hz)
T = 1.0  / Fs       # Sampling period (seconds)
A = 1.0         # Amplitude of sine-wave
f = 4.7           # Frequency of sine-wave (Hz)
y = [0]*80 ; x = [0]*80  # make x & y both lists
for n in range(0,80):
    y[n]=  A * numpy.sin( 2 * numpy.pi * f * n * T )
    x[n]=  n*T
fig, ax0 = pyplot.subplots(1) 
ax0.stem(x, y)
ax0.set_xlabel("Time (seconds)") ; ax0.set_ylabel("voltage")
ax0.grid(True)
fig, ax1 = pyplot.subplots(1)
ax1.step(x, y)
ax1.set_xlabel("Time (seconds)") ; ax1.set_ylabel("voltage")
```

## Section 9.8: Quantising sine-wave
In some applications we may have to consider using fewer bits per sample.  If we decided to use only 3 bits per sample rather than 16, we could only represent integers in the range -4 to +3.     The following code does not actually create true 3-bit integers and instead creates 16-bit integers where only the least significant three bits are used.  This is wasteful, but convenient.  To exploit the saving in bit-rate we could pack five 3-bit integers into a 16-bit word (wasting a bit), or sixteen 3-bit integers into three 16-bit words.  


```python
# Cell 2 for quantising y with 3 bits per sample (See Subsection 9.10)
y=numpy.array(y) ; x=numpy.array(x)
quant_y = numpy.round (y * 3.5 - 0.5 ) 
quant_y = numpy.int16(quant_y)
fig, ax2 = pyplot.subplots(1) 
ax2.stem(x, quant_y)
ax2.set_xlabel("Time (seconds)") ; ax2.set_ylabel("voltage")
ax2.grid(True)
fig, ax3 = pyplot.subplots(1)
ax3.step(x, quant_y)
ax3.set_xlabel("Time (seconds)") ; ax3.set_ylabel("voltage")
ax3.grid(True)
```

## Section 10: Speech and music files
The following code serves as a system test and also illustrates how quantisation may be achieved. 

```python
import numpy as np
from scipy.io import wavfile


# Read in a wav file from working directory
(Fs, speech) = wavfile.read("HQ-speech44100-mono.wav")


# Display sampling frequency (Hz) as read from the wav file:
print "Fs = ", Fs

SM=max(abs(speech))           # Get maximum amplitude
speech = speech/float(SM)    # Scale maximum to 1 (note the float)


# Plot graph of the whole speech with 16 bits per sample:
from matplotlib import pyplot as plt
% matplotlib inline	
fig,axs=plt.subplots(1)
axs.plot(speech)


# Plot 2000 samples of the speech with 16 bits per sample:
fig,axs=plt.subplots(1)
axs.plot(speech[20000:22000])


#Qspeech = np.round(speech * (2**4 -1))  #Quantise to +/-15 as float
Qspeech = np.round(speech * (2**4 -0.5) - 0.5)  #Quantise to -16 to +15 as float
Qspeech = np.int16(Qspeech)

# Plot 2000 samples of quantised speech: int16 using 5 bits per sample:
fig,axs=plt.subplots(1)
axs.plot(Qspeech[20000:22000])


# Increase amplitude of quantised speech to approx 32000 & plot it all:
#scaledQs=speech*(2**11)
scaledQs=np.int16( (speech+0.5)*2**11)
fig,axs=plt.subplots(1)
axs.plot(scaledQs)


# Plot 2000 samples of the scaled-up quantised speech:
fig,axs=plt.subplots(1)
axs.plot(scaledQs[20000:22000])


# Now write the 5-bit quantised speech to a wav file:
wavfile.write("quantisedSpeech5.wav",Fs,scaledQs)


# Listen to the quantised speech (from array):
from comp28512_utils import Audio
Audio(scaledQs,rate=Fs)


```
